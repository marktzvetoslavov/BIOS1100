## Volum til en sfære og Bergmanns regel

I 1847 beskrev den tyske biologen Carl Bergmann et mønster som siden har blitt oppkalt etter han. Bergmanns regel sier at dyr av enkelte dyrearter pleier å være større i kalde klima.

Bergmanns regel har flere unntak og regnes ikke som en naturlov, men tankegangen er enkel: Varme blir *tapt* proporsjonalt med overflaten, mens varme blir *generert* proporsjonalt med kroppsvolum. Med andre ord, en dobbel så stor overflate taper varme dobbelt så raskt, mens dobbelt kroppsvolum genererer dobbelt så mye varme.

I denne obligen vil vi se nærmere på Bergmanns regel, nærmere bestemt dette spørsmålet:

>Hva skjer med volum (produksjon av varme) og areal (tap av varme) når vi endrer størrelsen på et dyr? Hvordan påvirker disse endringene dyrets totale varmebalanse?

Lag et Python-program som regner ut og printer volumet til en sfære med radius $0.5m$. Det kreves ikke, men det er anbefalt at du bruker flere forskjellige variabler i programmet.


```python
# I følgende program skal vi lage et program som regner og printer ut volumet til en sfære med radius 0.5 meter.

import math

radius = float(0.5) 
volum = (4/3) * math.pi * (radius ** 3) # volum formelen 
print("Volument til en sfære med radius 0.5 meter gir", volum)
```

**Overflaten** til en sfære er gitt ved $4*\pi*r^2$. Utvid programmet slik at det printer både overflaten og volumet til en sfære med `radius` lik $0.5m$


```python
# Her skal vi følge samme prinsippp som i første oppgave. Selv om man trenger ikke å importere de samme tingene 2 ganger, 
# skal jeg gjøre dette for hele obligen, i tilfelle man bare øsnker å kjøre en celle om gangen. 

import math 
radius = float(0.5) 
volum = (4/3) * math.pi * (radius ** 3) # volum formelen 
overflate = 4 * math.pi * (radius ** 2) # overflate formel
print("En sfære med radius 0.5 meter gir", volum, "meter som volum og", overflate, "meter som overflate.")
```

Ved bruke samme program som i de to forrige deloppgavene, legg til (en) linje(r) som lagrer forholdet $\frac{Overflateareal}{Volum}$ avrundet til en desimal i en variabel og printer det.


```python
# Det vi skal gjøre i følgende oppgave er å lagre forholdet mellom overflate areal og volumet til sfæren. Vi skal 
# da avrunde svaret til en desimal i en variabel og printe det. Det som høres mest logisk for meg er å bruke 
# round funksjonen for å avrunde til en desimal. 

import math 
radius = float(0.5) 
volum = (4/3) * math.pi * (radius ** 3) # volum formelen 
overflate = 4 * math.pi * (radius ** 2) # overflate formel
forhold = round((overflate / volum), 1)
print("Forholdet mellom overflateareal og volum til en sfære med radius 0.5 meter gir", forhold, "meter. ")
```

Lag en løkke som kalkulerer forholdet $\frac{Overflateareal}{Volum}$ avrundet til en desimal og lagrer resultatene i en liste. Kjør programmet med disse radiusene:

$0.5, 1, 1.5, 2, 2.5, 3, 3.5$ og $4$ meter.


```python
# I følgende oppgave skal vi lage en løkke som kalkulerer forholdet mellom overflate og volum til en sfære. Det vi skal 
# gjøre i dette tilfelle er å bruke en liste med ulike radiuser og lagre svaret i et annet liste. Det jeg har valgt 
# å gjøre her er å løse den med 2 type løkker. Første svaret bruker en for-løkke og den andre bruker en while-løkke.

# NB: Dette er ikke en del av oppgaven, men allikevel velger å ta det med for å gjøre denne obligen mer spennende å jobbe
# med :D. Jeg har opprettet en funksjon som tar to lister som parametre og printer da forholdet til en sfære 
# gitt radiusen. 

def print_forhold(x, y):
    teller = 0
    for radius in x:
        print("Forholdet til en sfære med radius", radius, "m gir", y[teller], "m.")
        teller = teller + 1    
```

```python
# denne løsningen er implementert med en for-løkke. For å se at alt stemmer, kan du fjerne hashtagen som kaller på 
# print_forhold funksjonen. 
import math 
radiuser, forholder = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0], [] # radius liste + tomt liste for å legge forhold resultat.

for radius in radiuser: 
    forhold = (4 * math.pi * (radius ** 2)) / ((4/3) * math.pi * (radius ** 3))
    forholder.append(round(forhold, 1))

print_forhold(radiuser, forholder)
```

```python
# denne løsningen er implementert med en while-løkke. For å se at alt stemmer, kan du fjerne hashtagen som kaller på 
# print_forhold funksjonen.  
import math 
radiuser, forholder = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0], [] # radius liste + tomt liste for å legge forhold resultat.

x = 0
while x < len(radiuser):
        radius = radiuser[x]
        forhold = (4 * math.pi * (radius ** 2)) / ((4/3) * math.pi * (radius ** 3))
        forholder.append(round(forhold, 1))
        x = x + 1

print_forhold(radiuser, forholder)
```

**h)**
Hva får endringer i forholdet Overflateareal/Volum å si for dyrets evne til å holde på varmen og til å kjøle seg ned? Hvilken av disse er viktig i varme og kalde klima?
Og hvordan henger dette sammen med Bergmanns regel?



Vi kan tenke oss at med økende størrelse vil arealet øke proporsjonalt med kvadratet av lengden, mens volumet vil øke proporsjonalt med tredje potens av lengden. Med andre ord, dersom vi har to dyr som i utgangspunkt har samme form, der den ene er 10 cm langt og det andre er 100 cm, vil det dyret som er 100 cm være 10 ganger lengre enn det første dyret. Dette vil også føre til at det andre dyret har 100 ganger større overføate og 1000 ganger større volum. 

Med dette i tankene:
    - varmeproduksjon ---> proporsjonalt med volumet. (genereres i hele kroppen)
    - varmeutveksling ---> proporsjonalt med overflaten. (skjer gjennom overflaten)
 
Et større dyr vil derfor ha evnen til å holde varmen mye lengre og lettere, men derimot vanskeligere å kvitte seg med denne varmen. Et mindre dyr vil ha evnen til å bli kvitt varmen mye lettere, men vanskeligere å beholde den. Varme blir tapt proporsjonalt med overflaten, mens varme blir generert proporsjonalt med kroppsvolum. Med andre ord, en dobbel så stor overflate taper varme dobbelt så raskt, mens dobbelt kroppsvolum genererer dobbelt så mye varme. For mange typer dyr lønner det seg å være større i det kaldere klimaet, som naturligvis finnes på høye breddegrader og høyder. Mindre dyr har et mindre forhold mellom overflateareal og volum, noe som bidrar til å redusere varmetapet - et mønster kjent som Bergmanns regel. (Generelt har en mer massiv organisme et mindre overflateareal til volum.)
